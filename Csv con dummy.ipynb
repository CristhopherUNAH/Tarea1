```python
import pandas as pd
```

```python
datos = pd.read_csv('Informacion academica csv.csv', engine='python', index_col=0)
```

```python
datosDummyFirstName = pd.get_dummies(datos['FirstName'], prefix="FirstName")
datosDummyLastName = pd.get_dummies(datos['LastName'], prefix="LastName")
datosDummyNotas = pd.get_dummies(datos['Notas'], prefix="Notas")
```

```python
datos = pd.concat([datos, datosDummyFirstName], axis = 1)
datos = pd.concat([datos, datosDummyLastName], axis = 1)
datos = pd.concat([datos, datosDummyNotas], axis = 1)
```
